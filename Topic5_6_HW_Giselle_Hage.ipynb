```python
#Import library
import pandas as pd
import numpy as np

#Creating Dataframe
df1 = pd.DataFrame({
    "name": ["Anna", "Dane", "David", "Kevin", "Joe","Rose"],
    "weight": [45.2,50.5,62.1,64.0,59.0,np.nan],
    "height": [155,160,162,170,167,159],
    "birth_year":[2000,1997,1996,1987,1998,2003],
    "gender": ["female","male","male","male","male","female"],
    "city": ["New York","Chicago","Austin","New York",np.nan,"New York"]
    })
#Replacing nan as float type in the weight column
df1 = df1.replace("weight", np.nan).astype({'weight': float})
df1
```

```python
#Creating another dataframe
df2 = pd.DataFrame({
    "nick_name": ["Anna", "David", "Dane","Rose"],
    "full_name": ["Anna Lewis", "David John", "Dane Sebastian","Rose Kimberly"]
    })
df2
```

```python
#Merging dataframes
a=df1.merge(df2, how = "left", left_on = "name", right_on = "nick_name")
a
```

```python
#Dropping nick_name column from the merged dataframe
a=a.drop('nick_name',axis=1)
a
```

```python
#Filling full_name column as name from earlier dataframe
a["full_name"]=a["full_name"].dropna()
a["full_name"]=a["full_name"].fillna(df1["name"])
a
```

```python
#Calculating bmi
b=a["weight"]
c=a["height"]/100
c=c*c
bmi=b/c
a=a.assign(BMI=bmi)
a
```

```python
#Assign class column depends on BMI values
x1 = np.where(a['BMI'] < 18,1,0)
x2 = np.where((a['BMI'] < 25) & (a['BMI']>=18.5),1,0)
x3 = np.where((a['BMI'] < 30) & (a['BMI']>=25),1,0)
x4 = np.where((a['BMI'] < 35) & (a['BMI']>=30),1,0)
x5 = np.where((a['BMI'] < 40) & (a['BMI']>=35),1,0)
x6 = np.where(a['BMI']>=40,1,0)
```

```python
for i in range(5):
    if x1[i]==1:
        a['class']='Underweight'
    if x2[i]==1:
        a['class']='Normal weight'
    if x3[i]==1:
        a['class']='Overweight'
    if x4[i]==1:
        a['class']='Obesity class 1'
    if x5[i]==1:
        a['class']='Obesity class 2'
    if x6[i]==1:
        a['class']='Obesity class 3'
a
```
