### Jawaban No. 3
Langkah pengerjaan:
1. Mengidentifikasi data
3. Mencari mean tertinggi

```python
df.isna().sum()
```

```python
q1 = df["Annual_Price"].quantile(0.25)
q3 = df["Annual_Price"].quantile(0.75)

iqr = q3-q1 #Interquartile range
fence_low  = q1-1.5*iqr
fence_high = q3+1.5*iqr
df.loc[(df["Annual_Price"] < fence_low) | (df["Annual_Price"] > fence_high)]
```

```python
df_clean = df.loc[(df["CleanedPrice"] >= fence_low) & (df["CleanedPrice"] <= fence_high)]

df_clean
```

```python
regionmean = df.groupby("Region").agg({"CleanedPrice": "mean"})
regionmean.plot(kind='bar')
```
